#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot is diagram where each data is represented by dot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(4,2))
plt.scatter(x,y,c='orange')
plt.title('X versus y scatter plotting')
plt.xlabel('This is x-axis')
plt.ylabel('This is y-axis')
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar plot is type of data visualization used to represent the data in rectangular form.
its easy to understand and also used to when we are showing segment between two groups.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color="orange")
```

```python
plt.barh(company,profit,color="orange")
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum

Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.show()
```

```python
plt.boxplot(box2)
plt.show()
```
